## Multiple plots by superimposition

```python
#We will compare cosine and sin curve here
X = np.linspace(0,2*np.pi,101)
y = np.sin(X)

z = np.cos(X)


xticks = np.linspace(0,8,15)
yticks = np.linspace(-1,1,11)
plt.plot(X,y,'r',X,z,'k')
plt.xlabel("Angle in radian")
plt.ylabel("Magnitude")
plt.title("Sign curve")
plt.xticks(xticks)
plt.yticks(yticks)
plt.legend(['Sin','Cosine'])
plt.grid()
plt.axis([0.0,6.5,-1.1,1.1]) # 4 values has to be passed
plt.show()
```

### How to move X and Y axis in matplotlib

```python
X = np.linspace(-2.0*np.pi,2.0*np.pi,201)
y = np.sin(X)

z = np.cos(X)

plt.plot(X,y,'r',label='Sine')
plt.plot(X,z,'g',label='cosine')

plt.axes().spines['bottom'].set_position(('data',0)) # pushing the x asis
plt.axes().spines['left'].set_position(('data',0)) #Pushing the y axis

#Add a horizontal line at the bottom
#plt.axhline(y=-1)
plt.axhline(y = plt.ylim()[0],color='k') # Y limit gives (ymin,ymax) #putting a horizontal line
plt.axvline(x = plt.xlim()[0],color='k') # Y limit gives (ymin,ymax) #putting a vertical line

plt.legend()
```

# Pie Charts 

```python
plt.style.use("default")
distros = [
    'Arch Linux','Ubuntu','Debian','Fedora','Linux Mint',
    'Xubuntu','Kubuntu','Gentoo','OpenSUSE','Other'
]

colors = [
    'blue','orange','red','purple','green','chocolate',
    'beige','cyan','magenta','grey'
]


users_data = np.array([881,697,262,209,117,99,97,58,366,222])

#To make your own auto percentage
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}% ({v:d})'.format(p=pct,v= val)
    return my_autopct

explodes = np.zeros(10)
explodes[0] = 0.3

plt.pie(users_data,
        labels=distros,
        colors=colors,
        autopct=make_autopct(users_data),
        radius=1.3,
        center=(0,0),
        startangle=90,
        rotatelabels=5,
       explode=explodes,
        pctdistance=0.7,
       labeldistance=1.1,
       shadow=True,
       frame=False)
plt.axis("equal") # To make the plot circular
plt.show()
```

# How to annotate plots - IMPORTANT!!!!!

```python
plt.style.use("default")
X = np.linspace(0,2.0*np.pi,101)

c , s = np.cos(X), np.sin(X)

plt.plot(X,c,color='r',label='cosine')
plt.plot(X,s,color='g',label='Sine')

#Suppose we want to plot the maxima here,Sine maxima is in pie /2
plt.plot(0.5*np.pi,np.sin(0.5*np.pi),'bo')
#How to annotate this

plt.annotate(s='Sine local maxima',
             xy=(0.5*np.pi,np.sin(0.5*np.pi)),
            xytext=(1.0*np.pi,  0.75),
            arrowprops = dict(facecolor='blue',shrink=1.0))
plt.plot()
plt.show()
```

# Subplots using GridSpec

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
```

```python
fig1 = plt.figure(1,figsize=(4,4))
gs1 = gs.GridSpec(nrows=2,ncols=2)

#axis handle for plot 1
ax1 = plt.subplot(gs1[0,0])
ax1.text(x=0.5 , y = 0.5 , s='ax1' , va ='center' , ha='center')


#axis handle for plot 1
ax1 = plt.subplot(gs1[0,1])
ax1.text(x=0.5 , y = 0.5 , s='ax2' , va ='center' , ha='center')


#axis handle for plot 1
ax1 = plt.subplot(gs1[1,0])
ax1.text(x=0.5 , y = 0.5 , s='ax3' , va ='center' , ha='center')



#axis handle for plot 1
ax1 = plt.subplot(gs1[1,1])
ax1.text(x=0.5 , y = 0.5 , s='ax4' )


plt.suptitle("Subplot via GridSpec")
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
#Trying to do some complex grid structure using GridSpec

fig2 = plt.figure(2,figsize=(4,4))

gs2 = gs.GridSpec(nrows=3,ncols=3)

ax1 = plt.subplot(gs2[0,:])
ax1.text(x=0.5 , y = 0.5 , s='spanning to three column' , va ='center' , ha='center')


ax2 = plt.subplot(gs2[1,0:2])
ax2.text(x=0.5 , y = 0.5 , s='spanning to two column' , va ='center' , ha='center')

ax3 = plt.subplot(gs2[1:,2])
ax3.text(x=0.5 , y = 0.5 , s='ax3' , va ='center' , ha='center')


ax4 = plt.subplot(gs2[2,0])
ax4.text(x=0.5 , y = 0.5 , s='ax4' , va ='center' , ha='center')


ax5 = plt.subplot(gs2[2,1])
ax5.text(x=0.5 , y = 0.5 , s='ax5' , va ='center' , ha='center')




plt.suptitle("Subplot via GridSpec")
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

## Creating multiple Grid Spec object in a single figure

```python
from matplotlib import gridspec
gs1 = gridspec.GridSpec(3, 3)
gs1.update(left=0.05, right=0.48, wspace=0.05)
ax1 = plt.subplot(gs1[:-1, :])
ax2 = plt.subplot(gs1[-1, :-1])
ax3 = plt.subplot(gs1[-1, -1])

gs2 = gridspec.GridSpec(3, 3)
gs2.update(left=0.55, right=0.98, hspace=0.05)
ax4 = plt.subplot(gs2[:, :-1])
ax5 = plt.subplot(gs2[:-1, -1])
ax6 = plt.subplot(gs2[-1, -1])
```
