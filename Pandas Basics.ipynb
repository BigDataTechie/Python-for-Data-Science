## Concat and merge explained

```python
np.arange(1,6)
```

```python
df1 = pd.DataFrame({'member_id' : np.arange(1,6) , 
                    'first_name' :  ['Adam','Adrian','Alan','Alexzander','Andrew'],
                    'last_name' : ['Abraham','Allan','Alsop','Anderson','Arnold']
                   })
```

```python
df1
```

```python
df2 = pd.DataFrame({'member_id' : np.arange(4,9) , 
                    'first_name' :  ['Benjamin','Blake','Boris','Brandon','Brian'],
                    'last_name' : ['Baker','Ball','Bell','Belly','Black']
                   })
```

```python
df2
```

# Concatenate Two dataframes

```python
#row  concatenate

pd.concat([df1,df2],axis=0,ignore_index=True)

```

```python
pd.concat([df1,df2],axis=1)
```

```python
#How the merging occurs
```

```python
#inner join


pd.merge(df1,df2,how='inner',on='member_id')
```

```python
pd.merge(df1,df2,how='left',on='member_id')
```

```python
pd.merge(df1,df2,how='outer',on='member_id',suffixes=('_left','_right'))
```

```python
pd.merge(df1,df2,how='left',on='member_id')
```
